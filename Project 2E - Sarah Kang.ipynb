## Introduction

### Scientific Question: Is there any genetic similarity between genes of the electric eel fish (Electrophorus electricus), the european eel (Anguilla anguilla) and platypi (Ornithorhynchus anatinus) who use electroreception? Specifically, are there overlap in genes related to excitation regulation? 

Electric eel fish, specifically of the Electrophorus electricus species, use their electric organs in a variety of vital functions ranging from communication, navigation, hunting, and defense. These electric organs have revealed significant differences in gene regulation between various genes related to cell excitation and excitation-contraction coupling. Both of these functions are involved in the electric changes (for example, action potential changes or calcium release within the cell) that are involved in many functions (Gallant et al., 2014).

In the study by Gallant, 5 main cellular functions of electric organs were identified and various genes per cellular function were measured for up or downregulation. Within these, the genes related to cell excitation included atp1a2a, atp1a3a, znrf2a, scn4aa, fgf13a. Genes aligned later will include atp1a2a, an ATPase for sodium/potassium transport, and fgf13a, a fibroblast growth factor.

In platypi, electric-based functions were the same as the electric eel's: communication, navigation, predation, and defense. (Traeger et al., 2017) Therefore, this leads us to hypothesize that these two species may have overlap in genes related to cellular excitation-- which is associated with electric changes on a cellular level, as previously identified. 

The european eel (Anguilla anguilla) is used as a control or constant, as the two eels should have similar genes related to excitation regulation. 

### Scientific Hypothesis: If the electric eel's electric organ cells and the platypus' bill electroreceptors use similar electroreception-related function, then there will be significant overlap of excitation genes in their genomic sequence (as well as between the electric eel and european eel). 

The genes of interest I am focusing on were taken from Gallant's article, who's study identified 23 total genes that were either positively or negatively involved in electric cellular function. Researchers used previously identified genes from other species of electric fish, as per previous studies, BLAST searched the genome of the electric eel to confirm existence, and finalized the list of genes they analyzed. I am using the genes identified from that study that were categorized as 'cell excitation' genes. 

To compare genes between the electric eel, european eel, and platypus, I located online datafiles for the DNA sequences of each species. Each of the eel's exctiation gene DNA sequence was acquired on NCBI, sourced from here: https://www.ensembl.org/Electrophorus_electricus/Info/Index. Similarly, gene DNA sequences for the european eel and platypus were sourced from NCBI: https://www.ncbi.nlm.nih.gov/gene.

The imported files below are FASTA format files that display raw DNA Sequences of the gene "atp1a2a" for the electric eel, platypus, and european eel, respectively. 

Below, I imported 3 of these files and converted each into a string in order to remove any excess spaces that prevented the following functions from reading the data. #### *Below is atp1a2a alignment*

```python
# converting sequence of electric eel's gene "atp1a2a" from an ncbi file into a string
eleceel = open("/Users/sarahkang/Downloads/atp1a2a_datasets/ncbi_dataset/data/gene.fna", "r")

atp1a2a_eel = ""

# changing the format of the file into a more legible string
for line in eleceel.readlines():
        if ">" not in line: 
            line = line.replace("\n","")
            atp1a2a_eel = atp1a2a_eel + line
            
# code check
print(atp1a2a_eel) 
```

```python
# converting sequence of platypus's gene equivalent of atp1a2a from an ncbi file into a string
platypus = open("/Users/sarahkang/Downloads/sequence.fasta", "r")
    
atp1a2_platypus = ""

# again changing the format of the file to be more legible
for line in platypus.readlines():
        if ">" not in line: 
            line = line.replace("\n","")
            atp1a2_platypus = atp1a2_platypus + line

# code check
print(atp1a2_platypus)
```

```python
# converting sequence of european eel's gene equivalent of atp1a2a from an ncbi file into a string
europeeel = open("/Users/sarahkang/Downloads/sequence (1).fasta", "r")

atp1a2a_europeeel = ""

# again changing the format of the file into a more legible string
for line in europeeel.readlines():
        if ">" not in line: 
            line = line.replace("\n","")
            atp1a2a_europeeel = atp1a2a_europeeel + line

# code check
print(atp1a2a_europeeel) 
```

## Part 3: Plotting the Results

To analyze the data from my pairwise sequence alignment from Part 2 (above), I used a heatplot. A heatplot will help to 2-dimensionally visually display the data of the gene alignments using varying color gradients as a representation of the magnitude of the numerical data (in this case, of alignment scores). 

Below, the heatplot is set up in a 3 by 3 grid with each axis containing the three species I compared for gene "atp1a2a", which is an ATPase for sodium/potassium transport. The data from Part 2 was loaded into a 3 by 3 matrix in the order I planned the heatplot to be visualized, then a variable with the dataset was created, and finally, a heatplot was generated using that data variable in a Seaborn package. Higher scores were correlated with lighter colors, while lower scores were correlated with darker colors, as the key on the right of the plot depicts. Because the grid of the heatplot visualizes some duplicates, the data and colors are repeated in the top left and bottom right corners.

Highest alignment scores were between a DNA sequence with itself, obviously, and the longer the DNA sequence, the higher the match and the higher the score; so, electric eel aligned with itself was the highest, visualized by the lightest tan color, followed by the european eel with itself in pink, and then the platypus with itself in the slightly darker pink. Then, the next highest score was the electric eel with the european eel, visualized by the purple, followed by the electric eel with the platypus in darker purple, and finally, the european eel with the platypus in black. This makes sense as the eels should have very similar DNA sequences as they are closely related to each other compared to the platypus. The platypus was more similar to the electric eel rather than european eel. Thus, my hypothesis about the electric eel and platypus having similarity between their cell excitation gene, atp1a2a, was correct.
